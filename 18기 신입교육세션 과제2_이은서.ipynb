## 객실위치

1. Cabin 위치와 생존률 비교
Cabin의 위치가 상층에 존재할수록(사고 위치로부터 떨어질수록) 생존률이 높을 것이다.
2. Cabin 앞의 알파벳이랑 Pclass 분포 비교
가설: 객실 등급이 높은 승객들이 상층 객실에 더 많이 존재하면 객실등급이 높은 승객일수록 생존률이 높을 것이다. 
3. Fare와 Pclass 비교
가설: Fare가 높은 객실일수록 Pclass가 높을 것이다. -> Fare가 높을수록 생존률 높을 것이다. 
4. Embarked(S,C,Q)와 Pclass의 관계 비교. 
Embarked(승선항구)별 Pclass 분포
가설: Embarked별 Pclass 분포가 다르다면, 객실등급이 높은 승객이 많이 탑승한 항구일수록 생존률이 높을 것이다. 

정리 이유: Cabin에 결측치가 많다. 채우려면 Fare + Pclass + Embarked 분석 후 랜덤으로 Cabin 지정

### 2. Cabin_location과 Pclass 분포 비교
(알고 있는 사실: Cabin의 위치가 상층에 존재할수록(사고 위치로부터 떨어질수록) 생존률 높을 것이다.)
가설: 객실 등급이 높은 승객들이 상층 객실에 더 많이 존재하면 객실등급이 높은 승객일수록 생존률이 높을 것이다. 

```python
# 범주형 vs 범주형 방법1
pd.crosstab(data1["Pclass"], data1["Cabin_location"])
```

```python
# 범주형 vs 범주형 방법 2
ct = pd.crosstab(data1["Pclass"], data1["Cabin_location"])
sns.heatmap(ct, annot=True, fmt="d", cbar=False)
plt.title("2x2 Contingency Table (Pclass x Cabin_location)")
plt.xlabel("Cabin_location")
plt.ylabel("Pclass")
# heatmap
```

<객실 위치>

객실이 상층부에 위치할수록 사고 위치와 멀어져있으며, 생존률이 높은 경향을 보인다는 것을 앞에서 확인하였다. 
분포표와 heatmap으로 확인한 결과 객실의 등급이 높을 수록 객실이 상층부에 존재한다는 사실을 알 수 있었다. 
그러므로 객실의 등급이 높을수록 생존률이 높을 것이라고 예측할 수 있다.

## Mr

```python
tmp = data1['Name_age'] == 'Mr' 
data1.loc[tmp, 'Age'] = ( 
    data1.loc[tmp] 
    .groupby('Pclass')['Age'] 
    .transform(lambda x: x.fillna(x.median())) 
)
```
