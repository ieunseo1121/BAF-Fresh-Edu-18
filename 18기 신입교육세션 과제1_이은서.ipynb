데이터명: lung_cancer

데이터 특징

출처: kaggle (https://www.kaggle.com/datasets/dhrubangtalukdar/lung-cancer-prediction-dataset)

분야: 의료

내용: 18~90세까지 총 5,000명의 자료가 포함되어 있으며, 폐암 위험을 분석하고 예측하기 위해 기록되었다. 변수는 개인적 특성과 의학적 요인을 반영하였으며, 먼저 개인적 특성은 연령, 성별, BMI, 생활습관, 환경적 요인 등으로 구성되어 있다. 의학적 요인은 흡연 횟수, 흡연 기간, 호흡기 건강 상태, 유전적 요인 등 폐암 결과에 영향을 미치는 것으로 알려진 요인으로 구성되어 있어 폐암 위험을 예측하는 데 유용하다. 또한 이 데이터를 통해 개인적 특성과 의학적 요인을 통합하여 다양한 위험 요인들이 어떻게 상호작용하여 폐암 위험에 영향을 미치는지 분석할 수 있다. 마지막으로 lung_cancer_risk는 여러 위험 요인을 종합해 폐암 고휘험군으로 분류되는지의 여부를 나타낸다. 하지만 이 데이터셋은 장기 추적 데이터가 아닌 정적 데이터로, 시간을 두고 기록하지 않았기 때문에 폐암의 원인을 흡연이라고 결정지을 수 없다는 단점이 있다. 

데이터명: 대전광역시_자치구별 신용카드(KB국민카드) 매출액

데이터 특징

출처: 공공데이터포털 (https://www.data.go.kr/data/15064213/fileData.do)

분야: 금융

내용: 2019~2020년 3월~7월까지의 대전광역시 자치구별 신용카드 매출액이 기록되어 있다. 연속형 변수로는 월별 매출이, 범주형 변수로는 자치구, 업종, 연도가 존재한다. 이 데이터를 통해 자치구별, 업종별 소비패턴을 분석할 수 있으며, 코로나가 유행하기 전인 2019년과 코로나 후의 2020년도를 비교해 코로나 전후 소비 변화를 분석할 수 있다. 또한 3월부터 7월까지 시계열 분석을 통해 시간적 흐름을 파악할 수도 있다. 이 데이터는 설문조사가 아닌 실제 결제 데이터를 기반으로 해서 현실 소비를 잘 반영했다는 장점이 있지만, 집계 기간이 짧으며 소비자들의 연령과 성별, 소득 등 소비패턴에 영향을 미치는 변수들이 반영되어있지 않다는 단점이 있다.
